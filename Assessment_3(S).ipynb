### 9) Released Movie Over the years?

```python
movies = netflix[netflix['type'] == 'Movie']

# Group by 'release_year' and count the number of movies
movies_per_year = movies.groupby('release_year').size().reset_index(name='count')

# Display the result
print(movies_per_year)

# Optional: Visualize the trend
plt.figure(figsize=(12, 6))
plt.plot(movies_per_year['release_year'], movies_per_year['count'], marker='o', linestyle='-')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

### 10) Movies made on year basis?

```python
# Filter for movies only
movies = netflix[netflix['type'] == 'Movie']

# Group by 'release_year' and count the number of movies
movies_per_year = movies.groupby('release_year').size().reset_index(name='count')

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='release_year', y='count', data=movies_per_year, marker='o', color='b')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
# Alternatively, create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='release_year', y='count', data=movies_per_year, palette='viridis')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

### 13 Show only the title of all tv shows that were released in india only

```python
india_tv_shows = netflix[(netflix['type'] == 'TV Show') & (netflix['country'] == 'India')]

# Extract the titles
titles = india_tv_shows['title']

# Create a horizontal bar plot
plt.figure(figsize=(10, len(titles) * 0.5))  # Adjust figure size based on the number of titles
sns.barplot(x=[1] * len(titles), y=titles, palette='viridis', orient='h')
plt.title('TV Shows Released in India')
plt.xlabel('')  # Remove x-axis label
plt.ylabel('Title')  # Label for y-axis
plt.xticks([])  # Remove x-axis ticks
plt.show()
```

### 14) show Top 10 Directors who gave the highest number of tv shows and movies to netflix?

### 15) in how many movies / tv shows , tom cruise was cast?

```python
tom_cruise_movies_shows = netflix[netflix['cast'].str.contains('Tom Cruise', case=False, na=False)]

# Count the number of movies/TV shows
count = tom_cruise_movies_shows.shape[0]

# Display the result
print(f"Tom Cruise was cast in {count} movies/TV shows.")
```

### 16) how many movies got the Tv-14 rating in the canada?

```python
# Filter for movies with TV-14 rating released in Canada
tv14_canada_movies = netflix[(netflix['type'] == 'Movie') & 
                             (netflix['rating'] == 'TV-14') & 
                             (netflix['country'] == 'Canada')]

# Count the number of movies
count = tv14_canada_movies.shape[0]

# Display the result
print(f"Number of movies with TV-14 rating released in Canada: {count}")
```

```python
rating_counts = netflix[(netflix['type'] == 'Movie') & (netflix['country'] == 'Canada')]['rating'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index,rotatelabels = True,  autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Distribution of Movie Ratings in Canada')
plt.show()
```
